## <b>A SIMPLE GAME USING NUMPY RANDOM FUNCTION AND PANDAS</b>

<b>IMPORT THE MODULES</b>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy import random
```

<b>READ YOUR STEP</b>

```python
df = pd.read_csv("datasource/dataset.csv")
```

<b>DATA WRANGLING</b>

```python
df.dropna(inplace = True)
```

<b>UNDERSTAND YOUR DATA</b>

```python
df.describe()
```

```python
data = df.loc[df['text'].str.contains('Why') | df['text'].str.contains('What')]
data
```

```python
data.groupby(['humor']).count()
data
```

```python
new_array = data.to_numpy()
new_array
```

<b>CLASSIFY DATA</b>

```python
true_array = []
false_array = []
my_array = []
for i in new_array:
    if i[-1] == True:
        true_array.append(i[0])
    else:
        false_array.append(i[0])
        
my_array.append(true_array)
my_array.append(false_array)

my_array
```

<b>CHECK FOR HUMOR</b>

```python
my_joke = random.choice(random.choice(my_array, p=[len(true_array)/len(new_array), 1 - (len(true_array)/len(new_array))]))
if my_joke in true_array:
    print(f'{my_joke} IS HUMOROUS')
else:
    print(f'{my_joke} is NOT HUMOROUS')
```

<b>PREPARE DATA FOR VISUALIZATION</b>

```python
humor_count = sum(1 for _, label in new_array if label == True)
nonhumor_count = sum(1 for _, label in new_array if label == False)
labels = ["Hunorous", "Not Humorous"]
counts = [humor_count, nonhumor_count]
colors = ["green", "red"]
```

<b>VISUALIZE DATA</b>

```python
# Plot the results
plt.bar(labels, counts, color=colors)
plt.xlabel("Text Classification")
plt.ylabel("Count")
plt.title("CLASSIFICATION OUTCOMES")
plt.show()
```
